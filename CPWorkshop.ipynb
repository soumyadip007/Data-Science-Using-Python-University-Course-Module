```python
#https://practice.geeksforgeeks.org/problems/find-all-pairs-whose-sum-is-x/0/

for _ in range(int(input())):
    n,m,x = map(int,input().split())
    dic1 = {k:1 for k in map(int,input().split())}
    dic2 = {k:1 for k in map(int,input().split())}
    cnt=0
    ar=[]
    for i in dic1.keys():
        if (x-i) in dic2:
            cnt+=1
            ar.append((i,x-i))
    ar.sort()
    if cnt==0: print(-1)
    else:
        for i in ar[:-1]:
            print(i[0],i[1],end=", ")
        print(ar[-1][0],ar[-1][1])
```
